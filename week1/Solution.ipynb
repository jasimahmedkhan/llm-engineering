```python
from openai import OpenAI
from IPython.display import display, Markdown
from dotenv import load_dotenv
from scraper import fetch_website_contents
import os

```

```python

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown. Write the summary in English!
"""

user_prompt = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.
"""
```

```python
load_dotenv(override=True)

OLLAMA_BASE_URL = os.getenv("OLLAMA_BASE_URL")

openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
def message_for(website):
    """Create message list for the LLMs"""
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website},
    ]
```

```python
def summarize(url):
    """Summarize the contents of a website"""
    contents = fetch_website_contents(url)
    messages = message_for(contents)
    response = openai.chat.completions.create(model="llama3.2", 
                                              messages=messages)
    return response.choices[0].message.content
```

```python
def display_summary(url):
    """Display the contents of a website"""
    response = summarize(url)
    display(Markdown(response))
    
```

```python
display_summary("https://www.google.com")
```

```python
display_summary("https://www.bbc.com")
```

## Google Gemini using Google AI Studio

```python
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

load_dotenv(override=True)

google_api_key = os.getenv("GOOGLE_API_KEY")

if not google_api_key:
    print("No API key was found - please be sure to add your key to the .env file, and save the file! Or you can skip the next 2 cells if you don't want to use Gemini")
elif not google_api_key.startswith(("AIz", "AQ")):
    print("An API key was found, but it doesn't start AIz")
else:
    print("API key found and looks good so far!")
```

```python
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)
response = gemini.chat.completions.create(model="gemini-2.5-flash-lite", 
                                          # messages=message_for("https://www.google.com"))
                                          messages=[{"role": "user", "content": "Tell me a fun fact"}])
display(response.choices[0].message.content)
```
